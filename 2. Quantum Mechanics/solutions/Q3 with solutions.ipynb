### آینه نیمه نقره‌اندود بدون اتلاف
آینه‌های نیمه نقره‌اندود را از دو منشور به هم چسبیده می‌سازند. از این آينه‌ها می‌توان برای تجزیه یا ترکیب دامنه‌های احتمال استفاده کرد. اگر حالت فوتونی سیستم در هر گام را با حالت زیر نشان دهیم
$$\vert \psi \rangle = \psi_0 \vert 0 \rangle + \psi_1 \vert 1 \rangle$$
که بردار $\vert 0 \rangle$ را به مسیرهای موازی A (مسیر شمال غرب به جنوب شرق) نسبت می‌دهیم و بردار $\vert 1 \rangle$ را به مسیرهای موازی B (مسیر جنوب غرب به شمال شرق) نسبت می‌دهیم.
آنگاه اثر این آينه را می‌توان به‌صورت زیر نشان داد.
$$\begin{pmatrix} \psi_d \\ \psi_c \end{pmatrix} = U \begin{pmatrix} \psi_a \\ \psi_b \end{pmatrix} = \begin{pmatrix} r_{00} & t_{01} \\ t_{10} & r_{11} \end{pmatrix} \begin{pmatrix} \psi_a \\ \psi_b \end{pmatrix}$$
که در رابطه بالا عملگر U باید یکانی بوده و نسبت به ورودی‌ها متقارن باشد. منظور از این رابطه این است که اگر یک فوتون در مسیر A حرکت کند و در حالت $\vert 0 \rangle$ باشد، پس از عبور از آینه نیمه نقره‌اندود در حالت 
$
r_{00}
\vert 0 \rangle
+
t_{10}
\vert 1 \rangle
$ قرار می‌گیرد. به این معنی که با دامنه 
$r_{00}$
فوتون بازتاب داده می‌شود و با دامنه
$t_{10}$
مسیرش را ادامه می‌دهد.

1.	توضیح دهید که این شرایط چرا باید برقرار باشند؟ 
2. 	این دو شرط چه قیدهایی روی ضرایب $r_{00}$، $r_{11}$، $t_{01}$ و $t_{10}$ ایجاب می‌کنند؟

### بررسی خروجی اندازه‌گیری‌ها
آزمایش روبرو را در نظر بگیرید که در آن از یک آینه نیمه نقره‌اندود استفاده‌شده است. عملگر متناظر با این آينه را می‌توان با ماتریس
$\frac{1}{\sqrt{2}} \begin{pmatrix} 1 & i \\ i & 1 \end{pmatrix}$
نشان داد.

1.	اگر فوتون‌ها یکی‌یکی در مسیر A تابیده شوند، نشان دهید که این فوتون‌ها با چه احتمالی در هر یک از خروجی‌های آشکارسازهای C و D پیدا می‌شوند.

### تداخل
ممکن است از قسمت قبل این نتیجه را بگیرید که این آینه هر بار که فوتون به آن می‌رسد به‌صورت تصادفی آن را در یکی از مسیرهای C و D می‌فرستند. این بخش و سوال آينده درصدد نقض این فرضیه است و نشان می‌دهد که این تعبیر کلاسیک نادرست است. آزمایش زیر را در نظر بگیرید: 

1. فوتونی در راستای A تابیده می‌شود. حساب کنید که C و D هرکدام با چه احتمالی فوتون را مشاهده می‌کنند. آیا می‌توان این پدیده را به‌صورت کلاسیکی هم توجیه کرد؟
راهنمایی: اثر آینه روی حالت‌ها را می‌توانیم با ماتریس زیر به دست آوریم. آینه همه چیز را برعکس می‌کند یعنی $\vert 0 \rangle$ را $\vert 1 \rangle $ می‌کند و برعکس.
$$U_{mirror} = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$$



2. حال فرض کنید جسم شفافی در یکی از مسیرها قرار دهیم که نقش آن ضرب شدن $e^{i \phi}$ به دامنه متناظر با آن مسیر باشد.این جسم شفاف متناظر با این ماتریس یکانی است
$$U_{\phi} = \begin{pmatrix} e^{i \phi} & 0 \\ 0 & 1 \end{pmatrix}$$
یعنی $U$ دامنه مسیر متناظر با 
$|0\rangle$
را در یک فاز ضرب می‌کند.
حال در آزمایش شکل زیر احتمال مشاهده فوتون در C و D را برحسب $\phi$ محاسبه کرده و نمودار آن را به ازای مقادیر مختلف $\phi$ رسم کنید.


## آماده‌سازی‌ها

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def dagger(x : str):
    return np.mat(x).conj().T
# e.g. dagger(np.mat([1 + 1j,0,0]))
```

```python
ket0 = np.array([[1], [0]])
ket1 = np.array([[0], [1]])

phi = np.pi / 4

U = np.array([[1, 1j],[1j, 1]]) / np.sqrt(2)
U_mirror = np.array([[0, 1],[1, 0]]) 
U_phi = np.array([[np.e**(1j * phi), 0],[0, 1]]) 
```

## جوابِ سؤال به شکلِ کد

```python
# first setup
psi_0 = ket0
psi_1 = np.matmul(U,psi_0)
print('psi_1 =', psi_1)
```

```python
measurement_C = np.array([[0,0],[0,1]])
measurement_D = np.array([[1,0],[0,0]])

p_c_1 = np.matmul(np.matmul(dagger(psi_1),measurement_C),psi_1)
p_d_1 = np.matmul(np.matmul(dagger(psi_1),measurement_D),psi_1)
print('P_C = ',p_c_1)
print('P_D = ',p_d_1)
```

```python
# second setup
psi_0 = ket0
psi_1 = np.matmul(U,psi_0)
psi_2 = np.matmul(U_mirror,psi_1)
psi_3 = np.matmul(U,psi_2)
print('psi_3 =', psi_3)
```

```python
p_c_2 = np.matmul(np.matmul(dagger(psi_3),measurement_C),psi_3)
p_d_2 = np.matmul(np.matmul(dagger(psi_3),measurement_D),psi_3)
print('P_C = ',p_c_2)
print('P_D = ',p_d_2)
```

```python
# third setup
psi_0 = ket0
psi_1 = np.matmul(U,psi_0)
psi_2 = np.matmul(U_mirror,psi_1)
psi_3 = np.matmul(U_phi,psi_2)
psi_4 = np.matmul(U,psi_3)
print('psi_4 =', psi_4)
```

```python
p_c_3 = np.matmul(np.matmul(dagger(psi_4),measurement_C),psi_4)
p_d_3 = np.matmul(np.matmul(dagger(psi_4),measurement_D),psi_4)
print('P_C = ',p_c_3)
print('P_D = ',p_d_3)
```

```python
n = 40
phi_s = np.pi * np.linspace(0.0, 2.0, num=n)

P_c = np.zeros(n)
P_d = np.zeros(n)

for i in range(n):
    U_phi = np.array([[np.e**(1j * phi_s[i]), 0],[0, 1]]) 
    psi_3 = np.matmul(U_phi,psi_2)
    psi_4 = np.matmul(U,psi_3)
    P_c[i] = np.real(np.matmul(np.matmul(dagger(psi_4),measurement_C),psi_4))
    P_d[i] = np.real(np.matmul(np.matmul(dagger(psi_4),measurement_D),psi_4))
    
```

```python
plt.plot(phi_s,P_c,label='$p_C$')
plt.plot(phi_s,P_d,label='$p_D$')
plt.legend()
plt.xlabel('$\phi$')
plt.show()
```

## جوابِ سؤال به شکلِ ریاضی

### آینه نیمه نقره‌اندود بدون اتلاف

#### جواب بخش اول
شرط یکانی بودن باعث می‌شود پیوستگی و عدم اتلاف احتمال کل را داشته باشیم. شرط تقارن هم برای ساختار متقارنی است که نسبت به ورودی‌ها دارد یعنی می‌خواهد انعکاس و عبورش متقارن باشد.
#### جواب بخش دوم
از شرط تقارن نسبت به ورودی ها این قید را داریم که:
$$t_{01} = t_{10} = t, \quad r_{00} = r_{11} = r$$
و از شرط یکانی بودن ماتریس اثر آینه قیدهای زیر را داریم:
$$\vert r \vert^2 + \vert t \vert^2 = 1, \quad rt^* + r^* t = 0$$
### بررسی خروجی اندازه‌گیری‌ها

#### جواب بخش اول
برای اینکه احتمال اینکه فوتون در هر یک از دو مسیر خروجی مشاهده شود را به دست آوریم ابتدا باید حالت سیستم بعد از عبور از آینه نیمه نقره‌اندود را به دست آوریم. بعد با داشتن این حالت می‌توانیم نتایج اندازه‌گیری روی آن را نیز به دست آوریم.

فرض می‌کنیم که حالت ورودی از مسیر A حالت زیر باشد:
$$\vert \psi_0 \rangle = \vert 0 \rangle $$
حالا برای اینکه حالت را بعد از عبور از آینه نیمه نقره‌اندود به دست آوریم کافی است عملگر آن را روی حالت ورودی به آن اثر دهیم. پس حالتی که از آن خارج می‌شود مطابق زیر است:
$$ \vert \psi_1 \rangle = U \vert \psi_0 \rangle = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & i \\ i & 1 \end{pmatrix} \begin{pmatrix} 1\\0\end{pmatrix} = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ i \end{pmatrix} = \frac{1}{\sqrt{2}} \vert 0 \rangle + \frac{i}{\sqrt{2}} \vert 1 \rangle = \frac{1}{\sqrt{2}} \vert D \rangle + \frac{i}{\sqrt{2}} \vert C \rangle$$
حالا می‌توانیم روی حالت بالا اندازه‌گیری را انجام دهیم. عملگرهای تصویر اندازه‌گیری به شکل زیر هستند:
$$ \hat{P}_C = \vert 1 \rangle \langle 1 \vert$$
$$ \hat{P}_D = \vert 0 \rangle \langle 0 \vert$$
پس احتمال اینکه فوتون در خروجی‌های C و D ظاهر شود به ترتیب برابر با مقادیر زیر است:
$$P(C) = \langle \psi_1 \vert \hat{P}_C \vert \psi_1 \rangle = \vert \langle \psi_1 \vert 1 \rangle \vert^2 = \frac{1}{2}$$
$$P(D) = \langle \psi_1 \vert \hat{P}_D \vert \psi_1 \rangle = \vert \langle \psi_1 \vert 0 \rangle \vert^2 = \frac{1}{2}$$

### تداخل

#### جواب بخش اول

در این بخش دوباره کاری که برای بخش قبل کردیم را انجام می‌دهیم. یعنی حالت سیستم را در هر مرحله حساب می‌کنیم. حالت‌های مختلف را به این شکل در نظر بگیرید:

0. حالت اولیه سیستم که از مسیر A وارد می‌شود. 
$$\vert \psi_0 \rangle  = \vert 0 \rangle$$
1. حالت بعد از عبور از اولین آینه نیمه نقره‌اندود.
$$\vert \psi_1 \rangle  = U \vert \psi_0 \rangle = U \vert 0 \rangle$$
2. حالت بعد از گذشتن از آینه‌ها.
$$\vert \psi_2 \rangle  = U_{mirror} \vert \psi_1 \rangle $$
3. حالت بعد از عبور از آینه نیمه نقره‌اندود دوم.
$$\vert \psi_3 \rangle  = U \vert \psi_2 \rangle $$

حالا حالت‌ها را مرحله به مرحله حساب می‌کنیم. $\vert \psi_1 \rangle$ را مطابق سوال قبل به صورت زیر داریم:
$$ \vert \psi_1 \rangle = U \vert \psi_0 \rangle = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & i \\ i & 1 \end{pmatrix} \begin{pmatrix} 1\\0\end{pmatrix} = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ i \end{pmatrix}$$

برای به دست آوردن حالت بعد از آینه ابتدا باید اثر آینه را به دست آوریم. آینه همه چیز را برعکس می‌کند یعنی $\vert 0 \rangle$ را $\vert 1 \rangle $ می‌کند و برعکس.
$$U_{mirror} = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$$
حالا می‌توانیم حالت بعد از آینه را به دست آوریم.
$$ \vert \psi_2 \rangle = U_{mirror} \vert \psi_1 \rangle = \frac{1}{\sqrt{2}} \begin{pmatrix} 0 & 0 \\ 1 & 0 \end{pmatrix} \begin{pmatrix} 1\\i\end{pmatrix} = \frac{1}{\sqrt{2}} \begin{pmatrix} i \\ 1 \end{pmatrix}$$
و در مرحله آخر حالت نهایی را به دست می‌آوریم.
$$ \vert \psi_3 \rangle = U \vert \psi_2 \rangle = \frac{1}{{2}} \begin{pmatrix} 1 & i \\ i & 1 \end{pmatrix} \begin{pmatrix} i\\1\end{pmatrix} = \begin{pmatrix} i \\ 0 \end{pmatrix}$$
و حالا با داشتن حالت نهایی مشابه سوال قبل می‌توانیم احتمال اینکه فوتون از مسیر C و D آمده باشد را به دست آوریم.
$$P(C) = \langle \psi_3 \vert \hat{P}_C \vert \psi_3 \rangle = \vert \langle \psi_3 \vert 1 \rangle \vert^2 = 0$$
$$P(D) = \langle \psi_3 \vert \hat{P}_D \vert \psi_3 \rangle = \vert \langle \psi_3 \vert 0 \rangle \vert^2 = 1$$
که این جواب را دیگر نمی‌توانیم با آن توجیه کلاسیکی که در بخش قبل به نظرمان می‌رسید توضیح دهیم.

#### جواب بخش دوم
این مساله تا جایی که حالت بعد از آینه و قبل از جسم شفاف جدید است کاملا مشابه بخش قبل است. پس فقط بعد از آن را می‌نویسیم.
3. حالت بعد از عبور از جسم شفاف.
$$\vert \psi_3 \rangle  = U_{\phi} \vert \psi_2 \rangle $$
4. حالت بعد از عبور از آینه نیمه نقره‌اندود دوم.
$$\vert \psi_4 \rangle  = U \vert \psi_3 \rangle $$

حالا باید مشخص کنیم که اثر جسم شفاف چیست. این جسم تنها دامنه مسیر بالایی را در عددی ضرب می‌کند. پس می‌توانیم به صورت زیر در نظرش بگیریم:
$$U_{\phi} = \begin{pmatrix} e^{i \phi} & 0 \\ 0 & 1 \end{pmatrix}$$

بنابراین خواهیم داشت:
$$ \vert \psi_3 \rangle = U_{\phi} \vert \psi_2 \rangle = \frac{1}{\sqrt{2}} \begin{pmatrix} e^{i \phi} & 0 \\ 0 & 1 \end{pmatrix} \begin{pmatrix} i\\1\end{pmatrix} = \frac{1}{\sqrt{2}} \begin{pmatrix} i e^{i \phi} \\ 1 \end{pmatrix}$$
$$ \vert \psi_4 \rangle = U \vert \psi_3 \rangle = \frac{1}{{2}} \begin{pmatrix} 1 & i \\ i & 1 \end{pmatrix} \begin{pmatrix} i e^{i \phi}\\1\end{pmatrix} = \frac{1}{{2}} \begin{pmatrix} i e^{i \phi} + i \\- e^{i \phi} + 1 \end{pmatrix}$$

که با ساده کردن حالت نهایی خواهیم داشت که:
$$ \vert \psi_4 \rangle = e^{\frac{i \phi}{2}} \begin{pmatrix} i cos(\frac{\phi}{2}) \\ sin(\frac{\phi}{2}) \end{pmatrix}$$

دوباره مثل دو بخش قبل می‌توانیم احتمال‌ها را به دست آوریم.
$$P(C) = \vert \langle \psi_1 \vert 1 \rangle \vert^2 = sin^2(\frac{\phi}{2})$$
$$P(D) = \vert \langle \psi_1 \vert 0 \rangle \vert^2 = cos^2(\frac{\phi}{2})$$
