# پارادوکس هاردی

در این مسأله به این سؤال می‌پردازیم که آیا می‌توان توصیفی موضعی از ذرات ارائه داد یا این توصیف با مکانیک کوانتومی در تناقض خواهد بود.

در مسألهٔ قبل با تقسیم کنندهٔ پرتو یا آینهٔ نیمه نقره اندود آشنا شدید. اگر R احتمال بازتاب از تقسیم کننده و T احتمال گذر باشد، در حالت کلی می‌توان عملگر کوانتومی متناظر با این تقسیم کننده‌ها را به صورت 

$$U = \bigl(\begin{smallmatrix}
\sqrt{T} & i\sqrt{R} \\ 
i\sqrt{R} & \sqrt{T} 
\end{smallmatrix}\bigr)$$

نشان داد.

در این مسأله به طور خاص همچنان با تقسیم کنندهٔ ۵۰:۵۰ سر و کار داریم یعنی $T=R=1/2$. شکل زیر را در نظر بگیرید: 

یک الکترون و یک پوزیترون در دو جهت عمود بر هم و به طور همزمان تابیده می‌شوند. در ادامه هر یک از این ذرات به یک تقسیم‌کننده می‌رسند. می‌دانیم که اگر الکترون و پوزیترون همزمان در یک مکان حضور داشته باشند، یکدیگر را نابود کرده و ۲ پرتو گاما تولید می‌کنند.

#### ۱)
در ابتدا فرض کنید که تقسیم‌کننده‌های پرتو ۵۰:۵۰ به صورت کلاسیک عمل کرده و هر بار که ذره‌ای به آن‌ها می‌رسد به صورت کاملا رندوم یا به مسیر مستقیم خود ادامه می‌دهد و یا در جهت عمود بازتاب می‌شود. در انتها، دو تا از آشکارسازها سیگنال می‌دهند و یا می‌دانیم که الکترون و پوزیترون یکدیگر را نابود کرده‌اند. احتمال سیگنال دادن هر زوج مرتبی از آشکارسازها را حساب کنید و به طور خاص نشان دهید به صورت کلاسیک احتمال سیگنال دادن همزمان U+ و U- برابر صفر است.

اگه وضعیت‌های کلاسیک رو با $e^{-}, e^{+}$ و وضعیت‌های کوانتومی رو با $ |e^{-}\rangle , |e^{+}\rangle$ نشون بدیم:

$ e^{-} \to \frac{1}{2}(v^{-} + u^{-}) \space ; \space e^{+} \to \frac{1}{2}(v^{+} + u^{+}) $

بعد از برخورد با آینه‌های نیمه‌نقره‌اندود: $ \Rightarrow e^{-}e^{+} = \frac{1}{4} \big( v^{-}v^{+} + v^{-}u^{+} + u^{-}v^{+} + u^{-}u^{+} \big)$ 

ولی یادتون باشه، اگه $e^{-}$ و $e^{+}$ به هم برسن, تبدیل به اشعه گاما می‌شن!
پس جمله‌ی $u^{-}u^{+}$ از دید آشکارگرهامون که فوتون تشخیص نمی‌دن $0$ هست و هیچ $e^{-}$ یا $  \space e^{+}$ای تشخیص داده نمی‌شه.

پس احتمالات تشخیص می‌شه: 
$$P(V^{-}V^{+}) = \frac{1}{4}$$ 

$$P(V^{-}U^{+}) = \frac{1}{4}$$

$$P(U^{-}V^{+}) = \frac{1}{4}$$ 

$$P(U^{-}U^{+}) = 0 \space $$ 

#### ۲)
حال در ادامهٔ مسأله بیایید فرض کنید تقسیم‌کننده‌ها به صورت کوانتومی عمل می‌کنند و حال دامنهٔ احتمال هر جفت از خروجی‌ها را بیابید. هم‌چنان فرض کنید:
$$|u^{+}\rangle|u^{-}\rangle \to |2\gamma\rangle$$
و نشان دهید احتمال دیدن همزمان U+ و U- برابر صفر است.(راهنمایی: وقتی الکترون به اولین تقسیم کننده می‌رسد، تابع موج آن به این شکل تغییر می‌کند: $|e^{-}\rangle \to 1/\sqrt{2} (|v^{-}\rangle +i|u^{-}\rangle)$)

یادتون باشه که همیشه اون جمله‌ای که درباره‌ی بازتاب شدنه ضریب i می‌گیره.

در زمان (t=0), داریم: $ |e^{-}\rangle \otimes |e^{+}\rangle $

بعد از برخورد با آینه‌ها، این استیت تبدیل می‌شه به:

$$ |e^{-}\rangle \to \tfrac{1}{\sqrt{2}}( |v^{-}\rangle + i|u^{-}\rangle) \space  \\ |e^{+}\rangle \to \tfrac{1}{\sqrt{2}}( |v^{+}\rangle + i|u^{+}\rangle)
$$

$$ |e^{-}\rangle \otimes |e^{+}\rangle \to \tfrac{1}{\sqrt{2}} \big( |v^{-}\rangle + i|u^{-}\rangle) \otimes \tfrac{1}{\sqrt{2}}( |v^{+}\rangle + i|u^{+}\rangle \big) \\ = \tfrac{1}{2} \big( |v^{-}v^{+}\rangle +  i|v^{-}u^{+}\rangle + i|u^{-}v^{+}\rangle - |u^{-}u^{+}\rangle \big) \\ \to \tfrac{1}{2} \big( |v^{-}v^{+}\rangle +  i|v^{-}u^{+}\rangle + i|u^{-}v^{+}\rangle - |2\gamma\rangle \big) $$

و باز هم، می‌بینیم که در عبارت نهایی، هیچ جمله‌ی $|u^{-}u^{+}\rangle$ وجود نداره!

##### مهم:

جمله‌ی نهایی رو می‌شه به صورت: $ |\psi\rangle -\tfrac{1}{2}|2\gamma\rangle $ نوشت که داریم:

$$ |\psi\rangle = \tfrac{1}{2} \big( |v^{-}v^{+}\rangle +  i|v^{-}u^{+}\rangle + i|u^{-}v^{+}\rangle \big) $$

این وضعیت رو به یاد داشته‌باشید چون توی سوال‌های بعدی خیلی ازش استفاده می‌کنیم.

#### ۳)
حال آزمایش بالا را در نظر بگیرید و نشان دهید که سنسور C+ تنها در صورتی سیگنال می‌دهد که U- نیز سیگنال دهد.

می‌بینیم که:

نکته: از این سوال به بعد برای سادگی دیگه وضعیت دوگاما رو نمی‌نویسیم چون همیشه ثابته.

$$ |u^{+}\rangle = \tfrac{1}{\sqrt{2}} \big( |d^{+}\rangle + i|c^{+}\rangle \big) $$

$$ |v^{+}\rangle = \tfrac{1}{\sqrt{2}} \big( |c^{+}\rangle + i|d^{+}\rangle \big) $$

$$ |\psi\rangle \Rightarrow \frac{1}{2\sqrt{2}} \big( |v^{-}\rangle \otimes ( |c^{+}\rangle + i|d^{+}\rangle) + i|v^{-}\rangle \otimes ( |d^{+}\rangle + i|c^{+}\rangle) + i|u^{-}\rangle \otimes ( |c^{+}\rangle + i|d^{+}\rangle) \big) \\ =\frac{1}{2\sqrt{2}} \big( |v^{-}c^{+}\rangle + i|v^{-}d^{+}\rangle + i|v^{-}d^{+}\rangle - |v^{-}c^{+}\rangle + i|u^{-}c^{+}\rangle - |u^{-}d^{+}\rangle \big) \\ = \frac{1}{2\sqrt{2}} \big( 2i|v^{-}d^{+}\rangle + i|u^{-}c^{+}\rangle - |u^{-}d^{+}\rangle \big) $$

برای حس‌گر $C^{+}$ فقط یه جمله‌ی $ c^{+}u^{-}$ وجود داره!

۴) مشابه قسمت قبل، نشان دهید که حسگر C- تنها در صورتی می‌تواند الکترون را اندازه‌گیری کند که حسگر U+ پوزیترون را اندازه‌گیری کرده باشد.

$$ |u^{-}\rangle \to \tfrac{1}{\sqrt{2}} \big( |d^{-}\rangle + i|c^{-}\rangle \big) $$

$$ |v^{-}\rangle \to \tfrac{1}{\sqrt{2}} \big( |c^{-}\rangle + i|d^{-}\rangle \big) $$

$$ |\psi\rangle \Rightarrow \tfrac{1}{\sqrt{2}} \big( |c^{-}v^{+}\rangle + i|d^{-}v^{+}\rangle + i|c^{-}u^{+}\rangle - |d^{-}u^{+}\rangle + i|d^{-}v^{+}\rangle - |v^{-}v^{+}\rangle \big) \\ = \tfrac{1}{\sqrt{2}} \big(2i|d^{-}v^{+}\rangle + i|c^{-}u^{+}\rangle + |d^{-}u^{+}\rangle \big) $$

برای حس‌گر $C^{-}$ فقط یه جمله‌ی $ c^{-}u^{+}$ وجود داره!

۵) حال آزمایش بالا را در نظر بگیرید و پس از به دست آوردن احتمال‌های هر کدام از ۴ خروجی ممکن برای آشکارسازها، توضیح دهید که با در نظر گرفتن سؤالات قبلی چطور ممکن است احتمال C+ و C- ناصفر باشد؟

$$ |u^{-}\rangle \to \tfrac{1}{\sqrt{2}} \big( |d^{-}\rangle + i|c^{-}\rangle \big) $$

$$ |v^{-}\rangle \to \tfrac{1}{\sqrt{2}} \big( |c^{-}\rangle + i|d^{-}\rangle \big) $$

$$ |u^{+}\rangle = \tfrac{1}{\sqrt{2}} \big( |d^{+}\rangle + i|c^{+}\rangle \big) $$

$$ |v^{+}\rangle = \tfrac{1}{\sqrt{2}} \big( |c^{+}\rangle + i|d^{+}\rangle \big) $$

$$ |\psi\rangle \Rightarrow \big( ( |c^{-}\rangle + i|d^{-}\rangle) \otimes ( |c^{+}\rangle + i|d^{+}\rangle ) +  ( |c^{-}\rangle + i|d^{-}\rangle ) \otimes ( |d^{+}\rangle + i|c^{+}\rangle ) + ( |d^{-}\rangle + i|c^{-}\rangle ) \otimes ( |c^{+}\rangle + i|d^{+}\rangle ) \big) \\ = \tfrac{1}{4} \big( |c^{-}c^{+}\rangle + i|c^{-}d^{+}\rangle + i|d^{-}c^{+}\rangle - |d^{-}d^{+}\rangle + |c^{-}d^{+}\rangle + i|c^{-}c^{+}\rangle + i |d^{-}d^{+}\rangle - |d^{-}c^{+}\rangle + |d^{-}c^{+}\rangle + i|d^{-}d^{+}\rangle + i|c^{-}c^{+}\rangle - |c^{-}d^{+}\rangle \big) \\ = \tfrac{1}{4} \big( (1+2i)|c^{-}c^{+} + i|c^{-}d^{+}\rangle + (i-1)|d^{-}c^{+}\rangle + (2i-1)|d^{-}d^{+}\rangle \big) $$

دقت کنید که یه جمله‌ی $ |d^{-}d^{+}\rangle $ با احتمال $\frac{1}{16}$ پیدا می‌شه که توی حالت‌های قبلی نداشتیم!
این اتفاق به پارادوکس هاردی معروفه که توش می‌بینیم با استفاده از قوانین مکانیک کوانتوم، آینه‌های نیمه‌نقره‌اندودمون کاری رو انجام می‌دن که از لحاظ کلاسیک غیرممکن بود!
